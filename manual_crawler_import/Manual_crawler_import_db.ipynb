```python
import os
import pandas as pd
import xlrd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 100)


# https://docs.google.com/spreadsheets/d/1tUP1n_2IW1E7CckPle4GNhaFmPpl-uH3vDZc73_G6H0/edit#gid=879487022
filename = "Article_training_data.xlsx"
```

```python
xls = pd.ExcelFile(filename)
sheet_names = xls.sheet_names[:-1] # exclude 'Summary'
print(sheet_names)
```

# Test data source

```python
df1 = pd.read_excel(
    filename,
    engine="openpyxl",
    sheet_name=0
)
df1.head()[:3]
```

```python
# use melt() to pivot the columns Subject1, Subject2, and Subject3 into one column
df_melted = df1.melt(id_vars=['url'], 
                    value_vars=df1.columns, var_name='Subject', value_name='paragraph')

# display the result
df_filtered = df_melted.dropna()
df_sorted = df_filtered.sort_values(['url', 'Subject'], ascending=[True, True])
df_final = df_sorted.drop('Subject', axis=1)
df_final

```

# Loop pivot through all sheets and concat them

```python
dfs = []

for sheet_name in sheet_names:
    df1 = pd.read_excel(
        filename,
        engine="openpyxl",
        sheet_name=sheet_name
    )

    df_melted = df1.melt(id_vars=['url'], 
                        value_vars=df1.columns, var_name='Subject', value_name='paragraph')


    df_filtered = df_melted.dropna()
    df_sorted = df_filtered.sort_values(['url', 'Subject'], ascending=[True, True])
    df_sorted['category'] = sheet_name
    df_sorted['paragraph'] = df_sorted['paragraph'].str.strip()
    df_final = df_sorted.drop('Subject', axis=1)
    dfs.append(df_final)
```

```python
df_concat = pd.concat(dfs)
df_concat
```

```python
df_grouped = df_concat.groupby(['url','category']).agg({'paragraph': list})
df_grouped = df_grouped.reset_index()
df_grouped
```
