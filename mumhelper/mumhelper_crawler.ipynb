```python
import os
import json
import re
import time
import requests, psycopg2
import traceback
from bs4 import BeautifulSoup
```

```python
root_url = "https://mumhelper.com/"
data = requests.get(root_url, verify=False)
soup = BeautifulSoup(data.text)
content_soup = soup.find_all("h4", {"class": "homefeattitle"})

```

```python
category_link_urls = []
for c in content_soup:
    for link in c.find_all('a'):
        if 'feed' not in str(link.get('href')):
            link_str = str(link.get('href'))
            category_link_urls.append(link_str)
            
for c in category_link_urls:
    print("@@@", c)
```

```python
url_list = set()

for category_link_url in category_link_urls:

    data = requests.get(category_link_url, verify=False)
    soup = BeautifulSoup(data.text)
    page_soup = soup.find_all("div", {"class": "wp-pagenavi iegradient"})

    page_link_url = []

    for p in page_soup:
        for link in p.find_all('a'):
            link_str = str(link.get('href'))
            page_link_url.append(link_str)
    
    last_page_url = page_link_url[len(page_link_url)-1]
    last_page_component = last_page_url.strip("/").split("/")
    last_page = int(last_page_component[len(last_page_component)-1])
    
    for i in range(1,last_page+1):
        link_str = category_link_url + "page/" + str(i) + "/"
        data = requests.get(link_str, verify=False)
        soup = BeautifulSoup(data.text)
        
        p = soup.find_all("div", {"class": "post-thumb"})
        for pp in p:
            for link in pp.find_all('a'):
                link_str = str(link.get('href'))
                url_list.add(link_str)
```

```python
f = open("mumhelper_url.txt", "w")

f.write(str(url_list))
f.close()
```
