```python
import os
import json
import re
import time
import requests, psycopg2
import traceback
from bs4 import BeautifulSoup
```

```python
import time
# url_list = set()

f=open("mktoops_url.txt","w")

for i in range(2,1143):
  print(i)
  root_url = "https://www.marketingoops.com/page/"+str(i)+"/"
  data = requests.get(root_url)
  soup = BeautifulSoup(data.text)
  content_soup = soup.find("div", {"id": "content"})
  time.sleep(0.05)

  for link in content_soup.find_all('a'):
    link_str = str(link.get('href'))
    if '#' not in link_str and '?' not in link_str:
      url_list.add(link_str)

f.write(str(url_list))
f.close()
```
